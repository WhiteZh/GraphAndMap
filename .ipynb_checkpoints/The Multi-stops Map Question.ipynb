# The Multi-stops Map Question

## Problem Description
Let graph $ G $ be a connected graph, where set $ V_G $ contains all the vertices within graph $ G $ and set $ E_G $ contains all the edges within graph $ G $.
Let set $ V_S $ be a set of vertices, where $ V_S $ is a subset of set $ V_G $.
Let connected graph $ S $ be the smallest subgraph of graph $ G $ which contains every vertex within set $ V_S $.
<br>
Given set $ V_S $, what does one possible graph $ S $ look like?

## Representation Preparation

```python
# include required libraries
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import random
import math
```

```python
# define beautiful graph representation function
def show_graph_beautiful(G, size: int = 10):
    # position is stored as node attribute data for random_geometric_graph
    pos = nx.get_node_attributes(G, "pos")

    # find node near center (0.5,0.5)
    dmin = 1
    ncenter = 0
    for n in pos:
        x, y = pos[n]
        d = (x - 0.5) ** 2 + (y - 0.5) ** 2
        if d < dmin:
            ncenter = n
            dmin = d

    # color by path length from node near center
    p = dict(nx.single_source_shortest_path_length(G, ncenter))

    plt.figure(figsize=(size, size))
    nx.draw_networkx_edges(G, pos, alpha=0.4)
    nx.draw_networkx_nodes(
        G,
        pos,
        nodelist=list(p.keys()),
        node_size=size * 8,
        node_color=list(p.values()),
        cmap=plt.cm.Blues_r,
    )

    plt.xlim(-0.05, 1.05)
    plt.ylim(-0.05, 1.05)
    plt.axis("off")
    plt.show()
```

```python
# define graph representation function
def show_graph(G, with_weight = False, with_name = False, size: int = 10, highlight_nodes: list[int] = None, highlight_edges: list[tuple[int]] = None):
    highlight_nodes = highlight_nodes or []
    highlight_edges = highlight_edges or []

    pos = nx.get_node_attributes(G, "pos")
    plt.figure(figsize=(size, size))
    nx.draw_networkx_edges(G, pos, alpha=0.2)
    if with_weight:
        nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G, 'weight'), font_size=size / 3)
    nx.draw_networkx_nodes(
        G,
        pos,
        node_size=size * 4,
        node_color=[.4 for _ in G.nodes],
        cmap=plt.cm.Blues_r
    )
    if with_name:
        nx.draw_networkx_labels(G, pos=pos, labels={i: str(i) for i in list(G.nodes)}, font_size=size / 2)

    plt.xlim(-0.05, 1.05)
    plt.ylim(-0.05, 1.05)
    plt.axis("off")
    plt.show()
```

## Objects Generation

```python
# define the overall random seed
random_seed = 896803
```

```python
# create graph G
# Use seed when creating the graph for reproducibility
G: nx.Graph = nx.random_geometric_graph(15, 0.318, seed=random_seed)

# Assign weights based on Euclidean distance between vertices
for u, v in G.edges():
    x1, y1 = G.nodes[u]['pos']
    x2, y2 = G.nodes[v]['pos']
    weight = ((x1 - x2) ** 2 + (y1 - y2) ** 2) ** 0.5
    G.edges[u, v]['weight'] = weight
del u, v, weight, x1, y1, x2, y2
```

```python
# check the graph
show_graph(G, with_name=True, size=50, with_weight=True)
```

```python
# create set V_G
V_G: list[int] = list(G.nodes())
# check V_G
print(V_G)
```

```python
# create set V_S
random.seed(random_seed)
V_S: list[int] = random.sample(k=3, population=range(0, G.number_of_nodes()))
# check V_S
print(V_S)
```

## Solution Helpers

```python
# define class Path
class Path:
    def __init__(self, nodes: list[int], weight: float, attributes: dict = None):
        if len(nodes) == 0:
            raise ValueError('nodes should contain at least one node')
        self.nodes = nodes
        self.weight = weight
        self.attributes = attributes or {}

    @property
    def head(self):
        return self.nodes[-1]

    @property
    def root(self):
        return self.nodes[0]

    def __lt__(self, other: 'Path'):
        return self.weight < other.weight

    def __gt__(self, other: 'Path'):
        return self.weight > other.weight

    def __eq__(self, other: 'Path'):
        return self.weight == other.weight

    def __add__(self, other):
        node = other[0]
        weight = other[1]['weight']
        ret = Path(self.nodes + [node], self.weight + weight, self.attributes.copy())
        return ret

    def __str__(self):
        str_node = 'nodes: ' + str(self.nodes)
        str_weight = 'weight: ' + str(self.weight)
        return str_node + '\n' + str_weight

    def __getitem__(self, item):
        return self.attributes[item]

    def __setitem__(self, key, value):
        self.attributes[key] = value

    def reverse(self):
        self.nodes = self.nodes.reverse()
```

```python
# shortest path between 2 points algorithm
def closest_path(p1: int, p2: int, G: nx.Graph = G) -> Path:
    if p1 == p2:
        return Path(nodes=[p1], weight=0)

    tries: list[Path] = [Path(nodes=[p1], weight=0)]
    V_Dominated: dict[int, float] = {p1: 0}
    result: Path or None = None

    while len(tries):
        selected_try = tries[0]
        for each_try in tries:
            if each_try < selected_try:
                selected_try = each_try
        del each_try
        tries.remove(selected_try)
        if result is None or result > selected_try:
            for each_vid, each_attr in G[selected_try.head].items():
                each_weight = each_attr['weight']
                if each_vid not in V_Dominated or V_Dominated[each_vid] > each_weight:
                    this_try: Path = selected_try + (each_vid, each_attr)
                    V_Dominated[each_vid] = each_weight
                    if this_try.head == p2:
                        if result is None or result > this_try:
                            result = this_try
                    else:
                        tries.append(this_try)

    return result
```

```python
# generation of the simplest model algorithm
def drop_redundancy(G: nx.Graph = G, V_S: list[int] = V_S) -> nx.Graph:
    G_s: nx.Graph = nx.Graph()
    G_s.add_nodes_from([(i, {'pos': G.nodes[i]['pos']}) for i in V_S])

    paths: list[Path] = []

    for i, each_V_S in enumerate(V_S):
        for j, other_V_S in enumerate(V_S[i+1:], start=i+1):
            c_path: Path = closest_path(each_V_S, other_V_S, G=G)
            if not any(item in V_S for item in c_path.nodes[1:-2]):
                paths.append(c_path)

    G_s.add_edges_from([(path.head, path.root, {'weight': path.weight}) for path in paths])

    return G_s
```

## Solution Process

### Single path, no return?

```python
tries: list[Path] = [Path(nodes=[V_S[0]], weight=0, attributes={'achieved': [V_S[0]]})]
V_Dominated: dict[int, float] = {V_S[0]: 0}
candidate: Path or None = None

while not len(tries) == 0:
    selected_try = tries[0]
    for each_try in tries:
        if each_try < selected_try:
            selected_try = each_try
    tries.remove(selected_try)
    if candidate is None or candidate > selected_try:
        for each in G[selected_try.head].items():
            if each[0] not in V_Dominated or V_Dominated[each[0]] > each[1]['weight']:
                this: Path = selected_try + each
                this['achieved'] = this['achieved'].copy()
                if each[0] in V_S and each[0] not in this['achieved']:
                    this['achieved'].append(each[0])
                V_Dominated[each[0]] = each[1]['weight']
                if len(this['achieved']) == len(V_S):
                    if candidate is None:
                        candidate = this
                    elif candidate > this:
                        candidate = this
                else:
                    tries.append(this)

print(candidate)
```
